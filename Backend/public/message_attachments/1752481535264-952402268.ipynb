```python
import numpy as np 
import matplotlib.pyplot as plt
def tridiag(a1,a2,a3,n) :
        A=np.zeros(n)
        B=a1*np.ones(n)
        C=np.diagflat(B)
        D=a3*np.ones(n-1)
        E=np.diagflat(D,-1)
        F=a2*np.ones(n-1)
        J=np.diagflat(F,+1)
        A=C+E+J
        b=2*np.ones((n,1))
        return (A,b)
        
        
```

```python
import numpy as np

def matrice_diag_dominante(B):
    n = B.shape[0]
    #len(B)YAAtini nbre de ligne
    
    for i in range(n):
        diag_element = abs(B[i, i])
        row_sum = np.sum(np.abs(B[i, :])) - diag_element
        if diag_element <= row_sum:
            return False

    return True

#B = np.array([[3, -1, 1],
             # [2, 2, -1],
             # [1, -2, 6]])
```

```python
#B=tridiag(4,1,1,10)[0]
B=tridiag(1,1,1,10)[0]
matrice_diag_dominante = matrice_diag_dominante(B)


if matrice_diag_dominante:
    print("La matrice est à diagonale strictement dominante.")
else:
    print("La matrice n'est pas à diagonale strictement dominante.")

```

```python
def jacobi(B, b1, X0, epsilon):
        if matrice_diag_dominante(B)==True:
            D=np.diagflat(np.diag(B))
            E=-1*np.tril(B)-D#Jacobi
            F=-1*np.triu(B)-D#G_S
            M=D
            N=E+F
            k=0
            while np.linalg.rome(B.dot(X0)-b1)<epsilon :
                X0=np.linalg.inv(M).dot(N).dot(X0)+np.linalg.inv(M).dot(b1)
                k=k+1
            return (X0,k)
        return print("B n'est pas a diagonale strictement dominante")
```

```python
B1,b1=tridiag(4,1,1,10)
X0=np.ones((10,1))
epsilon=10**-6
jacobi(B1,b1,X0,epsilon)
```
